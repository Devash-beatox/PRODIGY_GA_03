```python
text = """
The quick brown fox jumps over the lazy dog. The dog barks at the fox.
The fox runs away quickly into the forest. The forest is dark and quiet.
The dog chases the fox through the forest. The fox hides behind a tree.
The dog stops and sniffs the air. The fox waits quietly behind the tree.
"""
```

```python
import random
from collections import defaultdict

def build_markov_chain(text, order=1):
    words = text.split()
    chain = defaultdict(list)
    for i in range(len(words) - order):
        key = tuple(words[i:i+order])
        next_word = words[i+order]
        chain[key].append(next_word)
    return chain

chain_order1 = build_markov_chain(text, order=1)
print("Sample of the chain (order 1):")
print(list(chain_order1.items())[:5])
```

```python
def generate_text(chain, order=1, length=30):
    key = random.choice(list(chain.keys()))
    result = list(key)
    for _ in range(length):
        next_words = chain.get(tuple(result[-order:]))
        if not next_words:
            break
        next_word = random.choice(next_words)
        result.append(next_word)
    return " ".join(result)

print("Order 1 generated text:")
print(generate_text(chain_order1, order=1, length=30))
```

```python
chain_order2 = build_markov_chain(text, order=2)
print("Order 2 generated text:")
print(generate_text(chain_order2, order=2, length=30))
```
